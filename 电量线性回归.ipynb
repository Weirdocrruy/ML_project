```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
data_train = pd.read_csv(r"E:/数据分析数据集/电量/2021电量预测(1).csv",encoding='gbk')
data_train
```

```python
data_test = pd.read_csv(r"E:/数据分析数据集/电量/2022电量预测.csv",encoding='gbk')
data_test
```

```python
X = data_train.iloc[:,1:11]
y = data_train.iloc[:,-1]
```

```python
# X_1 = StandardScaler().fit_transform(X)
linear = LinearRegression().fit(X,y)
linear.score(X,y)
```

```python
X_test = data_test.iloc[:,1:11]
# X_test_1 = StandardScaler().fit_transform(X_test)
predict = linear.predict(X_test)
predict[predict<0].shape
```

```python
data_test['y'] = predict
```
