```python
features=['地区','房间类型','新旧程度','总分','公司','出行住宿','校园生活']
data_feature = data[features]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
data_feature['地区'] = OrdinalEncoder().fit_transform(data_feature['地区'].values.reshape(-1,1)).astype('int')
data_feature['房间类型'] = OrdinalEncoder().fit_transform(data_feature['房间类型'].values.reshape(-1,1)).astype('int')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data_feature['新旧程度'] = LabelEncoder().fit_transform(data_feature['新旧程度'])
```

```python
# 归一化
from sklearn import preprocessing as pre
data_feature['总分'] = pre.scale(data_feature['总分'])
data_feature['公司'] = pre.scale(data_feature['公司'])
data_feature['出行住宿'] = pre.scale(data_feature['出行住宿'])
data_feature['校园生活'] = pre.scale(data_feature['校园生活'])
```

```python
data_feature.head()
```

```python
from sklearn import linear_model as LM
from sklearn.model_selection import train_test_split as tts
model = LM.LinearRegression()
y = data['房价']
x_train,x_test,y_train,y_test = tts(data_feature,y,test_size=0.3,random_state=40)
model.fit(x_train,y_train)
model.score(x_train,y_train)
```

```python
eps = np.mean(abs(model.predict(x_test)-y_test))
eps
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
model.score(x_train,y_train)
eps = np.mean(abs(model.predict(x_test)-y_test))
y_pre1 = model.predict(x_test)
```

```python
eps
```

```python
print('Coefficient:\n',model.coef_)
print('Intercept:\n',model.intercept_)
```

```python
from sklearn import svm
model = svm.SVC()
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
eps = np.mean(abs(model.predict(x_test)-y_test))
y_pre2 = model.predict(x_test)
eps
```

```python
from sklearn import svm
model = svm.SVR()
model.fit(x_train,y_train)
model.score(x_train,y_train)
```

```python
eps=np.mean(abs(model.predict(x_test)-y_test))
y_pre3=model.predict(x_test)
eps
```

```python
# 模型融合
def Model_ensembling(pred_y_list):
    ensembling_pred_y = np.array(pred_y_list).mean(axis=0)
    return ensembling_pred_y
```

```python
pred_y_list = [y_pre1,y_pre2,y_pre3]
```

```python
y_pre = Model_ensembling(pred_y_list)
```
