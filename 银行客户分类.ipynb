- 对分类型数据处理

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.iloc[:,1:15]
Y = data.iloc[:,-1]
```

```python
X
```

```python
Xtrain,Xtest,Ytrain,Ytest = train_test_split(X,Y,test_size=0.3,random_state=420)
```

```python
for i in [Xtest,Xtrain]:
    i.index = range(i.shape[0])
```

```python
xc = X.columns[X.dtypes == 'object'].tolist()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder().fit(Xtrain.loc[:,xc])
```

```python
oe.categories_
```

```python
Xtrain.loc[:,xc] = oe.transform(Xtrain.loc[:,xc])
Xtest.loc[:,xc] = oe.transform(Xtest.loc[:,xc])
```

```python
Xtrain.loc[:,xc].head()
```

```python
# 类别与数字对应
for d,j in enumerate(xc):
    print('\t')
    for i,n in zip(oe.categories_[d],range(len(Xtrain.loc[:,j].unique()))):
        print(i,n)
```

- 处理数值型变量

```python
xr = Xtrain.columns.tolist()
for i in xc:
    xr.remove(i)
xr
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
st = StandardScaler().fit(Xtrain.loc[:,xr])
Xtrain.loc[:,xr] = st.transform(Xtrain.loc[:,xr])
Xtest.loc[:,xr] = st.transform(Xtest.loc[:,xr])
```

```python
Xtrain.head()
```

```python
Xtest.head()
```

- 建立模型

```python
from sklearn.linear_model import LogisticRegression as LR
```

```python
lr = LR().fit(Xtrain,Ytrain)
lr.score(Xtest,Ytest)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC().fit(Xtrain,Ytrain)
svc.score(Xtest,Ytest)
```

- 进行独热编码

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
X
```

```python
xc = X.columns[X.dtypes == 'object'].tolist()
```

```python
Xc = X.loc[:,xc]
```

```python
Xr = Xtrain.columns.tolist()
for i in Xc:
    Xr.remove(i)
Xr
```

```python
enc = OneHotEncoder().fit(Xc)
result = enc.transform(Xc).toarray()
result
```

```python
newX
```

```python
names = enc.get_feature_names().tolist()+Xr
```

```python
newX.columns = names 
```

```python
newX.head()
```
